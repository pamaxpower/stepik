**Задание 5.\
Устройство состоит из трех деталей. Для первой детали вероятность выйти из строя в первый месяц равна 0.1, для второй - 0.2, для третьей - 0.25.\
Какова вероятность того, что в первый месяц выйдут из строя:\
a) все детали \
b) только две детали \
c) хотя бы одна деталь \
d) от одной до двух деталей?**

**Решение:**\
Для решения задач нам необходимо определить события и их вероятности:\
событие **A** - выход из строя первой детали, и его вероятность **P(A)=0.1**\
событие **A<sub>1</sub>** - событие, противоположное **A**, и его вероятность **P(A<sub>1</sub>)=1-P(A)=0.9** \
событие **B** - выход из строя первой детали, и его вероятность **P(B)=0.2**\
событие **B<sub>1</sub>** - событие, противоположное **B**, и его вероятность **P(B<sub>1</sub>)=1-P(B)=0.8** \
событие **C** - выход из строя первой детали, и его вероятность **P(C)=0.25**\
событие **C<sub>1</sub>** - событие, противоположное **C**, и его вероятность **P(C<sub>1</sub>)=1-P(C)=0.75** 

```python
A = 0.1
A1 = 0.9
B = 0.2
B1 = 0.8
C = 0.25
C1 = 0.75
```

a) Рассмотрим решение для 1 случая, когда "из строя вышли все детали".\
В этом случае полная вероятность **P** будет равна произведению всех вероятностей, а именно **A, B, C**:\
**Pa = A * B * C**

```python
Pa = A*B*C
print(f'Вероятность того, что все детали выйдут из строя равна: {Pa: .4f}')
```

b) Второй случай - когда "выйдут из строя две детали". Логично предположить, что третья останется исправной. В таком случае у нас возможны 3 сочетания:
- 1 и 2 сломаются, 3 нет: участвуют события **A, B, C<sub>1</sub>**
- 1 и 3 сломаются, 2 нет: события **A, C, B<sub>1</sub>**
- 2 и 3 сломаются, 1 нет: события **B, C, A<sub>1</sub>**

Вероятность каждого сочетания является произведением соответствующих событий, а полная вероятность второго случая будет суммой вероятностей событий:\
**Pb = A * B * C<sub>1</sub> + A * С * B<sub>1</sub> + B * С * A<sub>1</sub>**

```python
Pb = A*B*C1 + A*C*A1 + B*C*A1
print(f'Вероятность того, что две детали выйдут из строя равна: {Pb: .4f}')
```

c) Рассмотрим третий случай, когда "из строя выйдет хоть бы одна деталь". Этот случай включает в себя случаи, когда из строя может выйти одгна детали, две или сразу три, т.е. рассматривается случай противоположный случаю "все детали исправны".\
Его и рассмотрим. Вероятность того, что все детали исправны, будем вычислять по формуле:\
**P(0) = A<sub>1</sub> * B<sub>1</sub> * C<sub>1</sub>**\
Исходя из этого, вероятность третьего события будет равна: **Pc = 1 - P(0)**

```python
Pc = 1 - A1*B1*C1
print(f'Вероятность того, что из строя выйдет хотя бы одна деталь равна: {Pc: .4f}')
```

d) И последный, четвертый случай, когда "из строя выйдет одна или две детали".\
Из условия понятно, что это будет сумма двух вероятностей: **Pb** - вероятность выхода из строя двух деталей и **P(1)** - когда из строя вышла одна деталь.\
Чтобы посчитать вероятность выхода из строя одной детали обратимся к сочетаниям ,которые могут получиться в таком случае:
- 1 выйдет из строя, 2 и 3 нет: события **A, B<sub>1</sub>, C<sub>1</sub>**
- 2 выйдет из строя, 1 и 3 нет: события **B, A<sub>1</sub>, C<sub>1</sub>**
- 3 выйдет из строя, 1 и 2 нет: события **C, A<sub>1</sub>, B<sub>1</sub>**

По аналогии со вторым случаем ищем вероятность для 4 случая:\
**P(1) = A * B<sub>1</sub> * C<sub>1</sub> + B * A<sub>1</sub> * C<sub>1</sub> + C * A<sub>1</sub> * B<sub>1</sub>**\
Следовательно полная вероятность четвертого случая будет равна: **Pd = Pb + P(1)**

```python
P1 = A*B1*C1 + B*A1*C1 + C*A1*B1
Pd = Pb + P1
print(f'Вероятность того, что выйдет из строя от отдной до двух деталей равна: {Pd: .4f}')
```

Ответ:\
a) Выйдут из строя все детали: 0,5%\
b) Выйдут из строя две детали: 8%
c) Выйдет из строя хотя бы одна деталь: 46%
d) Выйдет из строя от одной до двух деталей: 45,5%
